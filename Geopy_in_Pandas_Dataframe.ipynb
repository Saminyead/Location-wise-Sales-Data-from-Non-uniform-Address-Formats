```python
#importing pandas and os module (os module is for filename directory reading)
import pandas as pd
import os
```

```python
#defining the path of our sales data directory
path = './SalesAnalysis/Sales_Data'
#show all files in the directory containing sales data files and storing them in a variable
files = os.listdir(path)
#defining an empty dataframe which will contain the sales data of all files
all_months_data = pd.DataFrame()
#loop for concatenating each index of the list files into one single dataframe
for file in files:
    #since there may be a seperate folder in the directory
    if path + '/' + file == path + '/Sales_Data_Out_Folder':
        continue
    #the main block of the for loop
    df = pd.read_csv(path + '/' + file)
    #concatenating each file into the var all_months_data
    all_months_data = pd.concat([all_months_data,df])
#reset index
all_months_data.reset_index()
```

```python
#This cell of code to clean the data of all 'NaN' rows
all_months_data = all_months_data.dropna()
#code to reset index for the new dataframe with no 'NaN'
all_months_data = all_months_data.reset_index()
#dropping the older index column from the individual sales data files
all_months_data =all_months_data.drop(columns=['index','level_0'])
all_months_data
```

```python
#this cell and following cell to calculate the total sale in each row (quantity*price each)
#code to get data types of each column
print(all_months_data.dtypes['Quantity Ordered'])
print(all_months_data.dtypes['Price Each'])
```

```python
#converting data into numeric form default string/object type
all_months_data['Quantity Ordered'] = pd.to_numeric(all_months_data['Quantity Ordered'],errors='coerce')
all_months_data['Price Each'] = pd.to_numeric(all_months_data['Price Each'],errors='coerce')
#new column for amount of sale
all_months_data['Sale Amount in USD'] = all_months_data['Quantity Ordered']*all_months_data['Price Each']
#all_months_data.loc[all_months_data['Quantity Ordered']=='Quantity Ordered']
print(all_months_data.dtypes['Price Each'])
print(all_months_data.iloc[510:520])
```

```python
#define a function to extract city and state names from text address using geopy
#import geopy
from geopy.geocoders import Nominatim
#function to get city and state name from string address using geopy
def get_city_state_from_address(address_str):
    #user_agent = API key, Nominatim is the geocoding API of OSM
    locator = Nominatim(user_agent="myGeocoder")
    #the string address to be geocoded
    #geocoding the string address to get the latitude and longitude
    location = locator.geocode(address_str)
    #concatenating the lat and long into a variable
    latlong = location.raw['lat'] + ", " + location.raw['lon']
    #using the reverse method to obtain latitude and longitude
    address_re = locator.reverse(latlong, language='en').raw
    #storing the address dictionary in 'address_in' variable
    address_in = (address_re.get('address'))
    #concatenating the 'city' and 'state' into one variable
    city_state = address_in['city'] + " " + address_in['state']
    return city_state
```

```python
#let us see the internal workings of the above function in the following two cells
locator = Nominatim(user_agent="myGeocoder")
location = locator.geocode(all_months_data['Purchase Address'][0])
#location.raw returns a dictionary containing the various feature of the geocoded address e.g. 'place_id', 'osm_id'
#including 'lat' and 'long' (print location.raw to see what the dictionary looks
print(location)
#concatenating the lat and long into a variable
latlong = location.raw['lat'] + ", " + location.raw['lon']
#reverse method takes the latlong and returns the detailed address in a format of a dictionary containing
#components like city, country etc
address_re = locator.reverse(latlong).raw
print(address_re)
```

```python
#loop going over each entry in Purchase Address column and fectching city and state
for p_address in all_months_data['Purchase Address']:
    p_addresses=get_city_state_from_address(p_address)
    print(p_addresses)
```
